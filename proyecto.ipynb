```python
import cv2 
import numpy as np
img = cv2.imread('D:\Artificial Vision Scripts\photos\ejercicio 1.jpg')
cv2.imshow('a' , img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
puntos = img.copy()
cv2.circle(puntos,(90, 800),10,(255,0,0),2)
cv2.circle(puntos,(900, 800),10,(255,0,0),2)
cv2.circle(puntos,(230, 530),10,(255,0,0),2)
cv2.circle(puntos,(900, 530),10,(255,0,0),2)
cv2.imshow('puntos' , puntos)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pts1 = np.float32([[160,630],[930,630],[85,835],[980,835]])
pts2 = np.float32([[0,0],
                   [1000,0],
                   [0,1000],   
                   [1000,1000]])
```

```python
pers = cv2.getPerspectiveTransform(pts1,pts2)
imgPers = cv2.warpPerspective(img, pers, (1000,1000))
cv2.imshow('pers' , imgPers)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
recta = imgPers.copy()
cv2.line(recta, (500,0), (500,1000),(255,0,0), thickness=1) # linea dado 2 puntos  thickness de 1 para 1 pixel de grosor
cv2.imshow('medio' , recta)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
