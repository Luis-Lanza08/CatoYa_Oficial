```python
import serial
import cv2 as cv
import numpy as np

#arduino = serial.Serial('COM7', 9600) # Reemplaza 'COM3' con el puerto en el que está conectado tu Arduino
capture = cv.VideoCapture(0) #guardar video en variable. Tambien se puede colocar 0  o 1 como parametros para leer webcam
constante = 1
i = 0
p_anterior = 0
control = 0

while True:
    
    isTrue, frame = capture.read() # hay que leer el video frame por frame (imagen por imagen)

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    threshold, thresh_inv = cv.threshold(gray, 100, 255, cv.THRESH_BINARY_INV )
    thresh_inv = cv.cvtColor(thresh_inv, cv.COLOR_GRAY2RGB)

    blancosEnFila = thresh_inv[100].astype(np.int32)
    x = np.arange(len(blancosEnFila))

    if(255 in blancosEnFila):
        bordes = np.where(blancosEnFila == 255)[0]


        cv.line(thresh_inv, (319,0), (319,500),(255,0,0), thickness=1)

        max = np.amax(bordes)
        min = np.amin(bordes)
        medio = (min+max)//2

        final = thresh_inv.copy()
        cv.line(final, (min,0), (min,500),(0,255,0), thickness=1) # linea dado 2 puntos  thickness de 1 para 1 pixel de grosor
        cv.line(final, (max,0), (max,500),(0,255,0), thickness=1) # linea dado 2 puntos  thickness de 1 para 1 pixel de grosor
        cv.line(final,(medio,0), (medio,500),(0,255,0), thickness=1)
        cv.imshow('Simple Thresholded Inverse', final)
        
        # PID
        p = np.zeros(len(x))
        vector_zeros = np.zeros(len(x))  # tamanio del verctor de error
        pesos = np.linspace(-((x.shape[0]/2)-1) , (x.shape[0]/2), num = x.shape[0]) # creando vector de error
        vector_zeros[medio] = constante #posicion de donde esta la camara

        p = pesos * vector_zeros # encontrar error 
        kp = 1

        i = i + p[medio]
        
        if ((p[medio]*i)<0):
            i = 0
        ki = 0.05

        d = p[medio] - p_anterior
        p_anterior = p[medio]
        kp = 0.1

        control = kp*p[medio] + ki*i + kp*d

        if control > 999:
            control = 1000
        elif control < -999:
            control = -1000

        print(control)
    else:
        cv.imshow('Simple Thresholded Inverse', thresh_inv)

    if cv.waitKey(20) & 0xFF== ord('d'):   # detener video con tecla 'd' y mostrar un frame cada 20 ms
        break
capture.release()
cv.destroyAllWindows()
```
